```python
import pandas as pd

# Convert Google Drive link to a direct download link
transaction = "https://drive.google.com/uc?id=1saEqdbBB-vuk2hxoAf4TzDEsykdKlzbF"

# Read the CSV file from the direct download link
dt = pd.read_csv(transaction)

# Display the last few rows of the dataframe
print(dt.tail())
```

```python
dt.isnull()
```

```python
dt.head()
```

```python
dt.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check if the dataset is loaded
if dt.empty:
    print("The dataset is empty. Please check the source.")
else:
    # Display histograms for numerical columns
    numerical_columns = dt.select_dtypes(include=['float64', 'int64']).columns
    if not numerical_columns.empty:
        dt[numerical_columns].hist(figsize=(10, 8), bins=20)
        plt.suptitle("Histograms of Numerical Features", fontsize=16)
        plt.show()
    else:
        print("No numerical columns found for histograms.")

    # Display a heatmap of correlations (if numerical data exists)
    if len(numerical_columns) > 1:  # At least two numerical columns are required for correlation
        plt.figure(figsize=(10, 6))
        sns.heatmap(dt[numerical_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
        plt.title("Correlation Heatmap")
        plt.show()
    else:
        print("Not enough numerical columns for a correlation heatmap.")

    # Plot a bar chart for the first categorical column (if available)
    categorical_columns = dt.select_dtypes(include=['object']).columns
    if not categorical_columns.empty:
        cat_col = categorical_columns[0]  # First categorical column
        plt.figure(figsize=(10, 6))
        dt[cat_col].value_counts().plot(kind='bar', color='skyblue')
        plt.title(f"Distribution of {cat_col}")
        plt.ylabel("Count")
        plt.xlabel(cat_col)
        plt.show()
    else:
        print("No categorical columns found for bar chart.")
```
